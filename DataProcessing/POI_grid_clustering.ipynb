### 法2 POI重分类

```python
poi_map={      # 居住类：0 交通与设施类：1 休闲类：2 办公与学校类：3 服务类：4
    '科教文化服务': 3,
    '公司企业': 3,
    '医疗保健服务': 4,
    '政府机构及社会团体': 3,
    '购物服务': 2,
    '风景名胜': 2,
    '生活服务': 4,
    '公共设施': 1,
    '体育休闲服务': 2,
    '餐饮服务': 2,
    '金融保险服务': 4,
    '汽车服务': 4,
    '摩托车服务': 4,
    '住宿服务': 0,
    '通行设施': 1,
    '商务住宅': 0,
    '汽车维修': 4,
    '交通设施服务': 1,
    '道路附属设施': 1,
    '汽车销售': 4,
    '事件活动': 4}
```

```python
# 使用映射字典将属性列中的值替换为对应的数字
# df['BASETYPE'] = df['BASETYPE'].map(value_to_number)
df['BASETYPE'] = df['BASETYPE'].map(poi_map)
df['BASETYPE'] = df['BASETYPE'].astype('object')
df
```

```python
# 使用 pivot_table 创建数据透视表，计算每个 GridID 中每种 BASETYPE 的计数
pivot_df = df.pivot_table(index='GridID', columns='BASETYPE', aggfunc='size', fill_value=0)
```

```python
# 获取 GridID 的所有可能值
all_grids = np.arange(489)

# 使用 reindex 方法确保数据透视表包含所有 GridID 的可能值，并填充缺失值为0
pivot_df = pivot_df.reindex(all_grids, fill_value=0)
pivot_df
```

```python
pivot_df.iloc[1]
```

```python
# 重新设置列索引，将多层级列转换为单层级列，并重命名列名
# pivot_df.columns = ['poi_count_' + str(col) for col in pivot_df.columns]
pivot_df.columns = ['poi_count_0', 'poi_count_1', 'poi_count_2', 'poi_count_3','poi_count_4']
pivot_df
```

```python
pivot_df['total_count'] = pivot_df.sum(axis=1) 
```

```python
# 将数据透视表保存为 CSV 文件
# csv_new = r"D:\系统默认\Documents\gradProject\rawData\data1\POI_grid_count.csv"
csv_new = r"D:\系统默认\Documents\gradProject\rawData\data1\POI_grid_count1.csv"
pivot_df.to_csv(csv_new, index=True)
```

```python
pivot_df_copy = pivot_df.copy()

# 添加总和计数列
pivot_df_copy['total_count'] = pivot_df_copy.sum(axis=1)  # 每行求和，得到总计数

# 计算每种 BASETYPE 的占比
for column in pivot_df_copy.columns[:-1]:  # 不包括最后一列（总计数列）
    pivot_df_copy[column] = pivot_df_copy[column] / pivot_df_copy['total_count']  # 计算占比

# 用0填补缺失值
pivot_df_copy = pivot_df_copy.fillna(0)

pivot_df_copy
```

```python
# 将数据透视表保存为 CSV 文件
# csv_new = r"D:\系统默认\Documents\gradProject\rawData\data1\POI_grid_percentage.csv"
csv_new = r"D:\系统默认\Documents\gradProject\rawData\data1\POI_grid_percentage1.csv"
pivot_df_copy.to_csv(csv_new, index=True)
```

### 前一种求percentage的方式并不合理，POI总数较少的特征无法体现，且列之间存在相互关联。因此改用poi/poi_type

```python
df_percentage = pivot_df.copy()
df_percentage.drop(columns=['total_count'])
```

```python
# 计算每个 BASETYPE 的总数
basetype_totals = df_percentage.sum()

# 将每个记录除以相应 BASETYPE 的总数，得到百分比
df_percentage = df_percentage.div(basetype_totals) * 100

df_percentage
```
